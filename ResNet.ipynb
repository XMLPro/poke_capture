## 使用するライブラリのimport

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

## オレンジ背景にアイコンを合成

```python
pokename = '788-0'
```

```python
overlay_img = cv2.imread('../YOLOv2/alpha/{}.png'.format(pokename), -1)
overlay_img = cv2.cvtColor(overlay_img, cv2.COLOR_BGRA2RGBA)
plt.imshow(overlay_img)
```

```python
for i in range(7):
    overlay_img = cv2.imread('../YOLOv2/alpha/{}.png'.format(pokename), -1)
    overlay_img = cv2.cvtColor(overlay_img, cv2.COLOR_BGRA2RGBA)
    overlay_img = cv2.resize(overlay_img, (128, 128))
    src_img = cv2.imread('../YOLOv2/backgrounds/{}.jpg'.format(i))
    src_img = cv2.cvtColor(src_img, cv2.COLOR_BGR2RGB)
    src_img = cv2.resize(src_img, (227,227))
    overlay_img = Image.fromarray(overlay_img)
    src_img = Image.fromarray(src_img)
    overlay_img = overlay_img.convert('RGBA')
    src_img = src_img.convert('RGBA')
    tmp = Image.new('RGBA', src_img.size, (255, 255, 255, 0))
    tmp.paste(overlay_img, (50, 50), overlay_img)
    result = Image.alpha_composite(src_img, tmp)
    result.save('./orange_data/{0}/{1}-{2}.png'.format(pokename, pokename, i), 'png')
```
